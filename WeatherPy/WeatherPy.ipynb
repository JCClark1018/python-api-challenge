```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# URL Building
url = "http://api.openweathermap.org/data/2.5/weather?q="
units = "imperial"
api = f"&units={units}&appid={weather_api_key}"
```

```python
# create lists
name = []
lat = []
lon = []
tempM = []
humidity = []
cloud = []
windS = []
country = []
dt = []

#Style
print ("Beginning Search")     
print ("-----------------------------")

#count variable
cx = 0 
cy = 0

#loop through cities, run api request, append lists
for city in cities:
    try:
        response = requests.get(url + city + api).json()
        print(f"Processing Record: {city}.")
        name.append(response['name'])
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        tempM.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        windS.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        cx += 1
        print (f"City: {city} processed. {cx} cities imported successfully")
        
    except:
        print(f"City: {city} not found. Skipping...")
        pass
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
wdict = {
    "City": name,
    "Lat": lat,
    "Lon": lon,
    "Max Temp": tempM,
    "Humidity": humidity,
    "Wind": windS,
    "Cloudiness": cloud,
    "Country": country,
}

weatherDF = pd.DataFrame(wdict)

print(f'{len(weatherDF)} cities added to the Dataframe')
weatherDF.head()
```

```python
weatherDF.to_csv('../Output/weather_data_df.csv', index=False)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#XY
NHLat = NH['Lat'] #x
NHMT = NH['Max Temp'] #y

#Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(NHLat, NHMT)
regress_values = NHLat * slope + intercept
line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
rsq = "R-Squared =" + str(round(rvalue**2,4))

#Plot
plt.scatter(NHLat, NHMT)

#line
plt.plot(NHLat,regress_values,"r-")
plt.annotate(line,(0,40),fontsize=15,color="red")
plt.annotate(rsq,(0,34), fontsize=15,color="red")

#label
plt.xlabel("Latitude")
plt.ylabel('Max Tempurature (F)')
plt.title('Northern Hemisphere Tempurature vs Latitude')

# Show plot
plt.show()

#save png
plt.savefig(f'../Output/NH.Temp.Lat.png')
```
